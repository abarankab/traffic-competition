```
def evaluate(boxes):
  results = {}
  signs = []
  
  for box in boxes:
    if box[IMAGE].size != 0:
      #print("TMP")
      label, probability = predict(box[IMAGE])
      if label != NONETYPE:
        if (results.get(label) == None or results.get(label)[0] < probability):
          results[label] = [probability, box[COORDINATES]]
  
  for label, meta in results.items():
    signs.append([label, meta[1]])
    
  return signs
```
