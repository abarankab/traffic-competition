```
OFFSET = 300

def pick_background():
    filename = random.choice([x for x in os.listdir('./landscapes')])
    img = cv.imread('./landscapes/' + filename, cv.IMREAD_UNCHANGED)
    c = random.randint(OFFSET, img.shape[0] - OFFSET)
    r = random.randint(OFFSET, img.shape[1] - OFFSET)
    w = h = 40
    img = img[c:c+w, r:r+h]
    img = gamma(img)
    img = noise(img)
    return img.astype('uint8')


def gamma(img):
    img_rgb = img[:, :, :3]
    if img.shape[-1] == 4:
        alpha = img[:, :, 3]
    else:
        alpha = np.full((40, 40, 1), 255)
    
    gamma = np.random.uniform(0.5, 1.5)
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
                      for i in np.arange(0, 256)]).astype("uint8")
    img_rgb = cv.LUT(img_rgb, table)
    return np.concatenate((img_rgb, alpha.reshape(40, 40, 1)), axis=-1)


def brightness(img):
    img1 = cv.cvtColor(img,cv.COLOR_RGB2HSV)
    random_bright = np.random.uniform(0.5, 2)  # TODO
    img1[:,:,2] = (img1[:,:,2]*random_bright) / 2
    img1 = cv.cvtColor(img1,cv.COLOR_HSV2RGB)
    if img.shape[-1] == 4:
        img1 = np.concatenate((img1, img[:, :, 3].reshape(40, 40, 1)), axis=-1)
    else:
        img1 = np.concatenate((img1, np.full((40, 40, 1), 255)), axis=-1)
    return img1.astype('uint8')


def noise(img):
    alpha = img[:, :, 3]
    noisy = img[:, :, :3]

    vals = len(np.unique(noisy)) / 2
    vals = 2 ** np.ceil(np.log2(vals))
    noisy = np.random.poisson(noisy * vals) / float(vals)
    return np.concatenate((noisy, alpha.reshape(40, 40, 1)), axis=-1)


def blur(img):
    return cv.medianBlur(img, 3)


def transform_img(img, ang_range, shear_range, trans_range):
    '''
    This function transforms imgs to generate new imgs.
    The function takes in following arguments,
    1- img
    2- ang_range: Range of angles for rotation
    3- shear_range: Range of values to apply affine transform to
    4- trans_range: Range of values to apply translations over.

    A Random uniform distribution is used to generate different parameters for transformation

    '''
    # Rotation

    ang_rot = np.random.uniform(ang_range)-ang_range/2
    rows,cols,ch = img.shape
    Rot_M = cv.getRotationMatrix2D((cols/2,rows/2),ang_rot,1)

    tr_x = trans_range*np.random.uniform()-trans_range/2
    tr_y = trans_range*np.random.uniform()-trans_range/2
    Trans_M = np.float32([[1,0,tr_x],[0,1,tr_y]])

    pts1 = np.float32([[5,5],[20,5],[5,20]])

    pt1 = 5+shear_range*np.random.uniform()-shear_range/2
    pt2 = 20+shear_range*np.random.uniform()-shear_range/2

    pts2 = np.float32([[pt1,5],[pt2,pt1],[5,pt2]])

    shear_M = cv.getAffineTransform(pts1,pts2)

    img = cv.warpAffine(img,Rot_M,(cols,rows))
    img = cv.warpAffine(img,Trans_M,(cols,rows))
    img = cv.warpAffine(img,shear_M,(cols,rows))

    img = brightness(img)

    return img
```

```
def genImage(name):
  img = cv.imread('./{}.png'.format(name), cv.IMREAD_UNCHANGED)
  if img.shape[-1] != 4:
    img = np.concatenate((img, np.full((40, 40, 1), 255)), axis=-1)
  if name[0] == 'p':
    img = img.astype('float32')
    img = cv.resize(img, (36, 36))
    img = cv.copyMakeBorder(img, 2, 2, 2, 2, cv.BORDER_CONSTANT, value=(0, 0, 0, 0))
  else:
    img = cv.resize(img, (40, 40))
    
  if name != 'none':
    overlay = noise(blur(transform_img(img, 10, 3, 3)))
  else:
    overlay = np.zeros((40, 40, 3,))
  background = pick_background()
  result = np.zeros(overlay.shape)
  for i in range(40):
    for j in range(40):
      if overlay[i][j][3] == 255:
        result[i][j] = overlay[i][j]
      else:
        result[i][j] = background[i][j]
  return result.astype('uint8')
```

```
def auto_canny(image, g = 0.33):
  med = np.median(image)
  low = (1 - g) * med
  upp = (1 + g) * med
  
  return cv.Canny(image, low, upp)
```
